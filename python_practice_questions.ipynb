# List Comprehension Practice Questions

1. You're analyzing weather data. Given a list of temperatures, filter out all temperatures below 15°C and create a new list using list comprehension. Input: [5, 20, 12, 25, 8, 30, 3, 18]

```python
mylist=[5,20,12,25,8,30,3,18]
newList=[]
newList1=[]
for i in mylist:
    if(i<15):
        newList.append(i)
    else:
        newList1.append(i)
print(newList)
print(newList1)
```

2. An e-commerce platform filters product names by length. Given a list of product names, create a list containing only names with more than 5 characters using list comprehension. Input: ['phone', 'keyboard', 'mouse', 'monitor', 'usb']

```python
l1=['phone','keyboard','mouse','monitor','usb']
l2=[]
for i in l1:
    if len(i)>5:
        l2.append(i)
print(l2)
```

3. A restaurant wants to calculate ingredient costs. Given quantities (in kg), calculate total cost where each kg costs as its square value. Use list comprehension. Input: [2, 3, 5, 1, 4]

```python
quantities=[2,3,5,1,4]

ing_cost=[x**3 for x in quantities]
print(ing_cost)
total_cost=sum(ing_cost)
print(total_cost)
```

4. A password validator extracts uppercase letters to check password strength. Given a string, extract all uppercase letters using list comprehension. Input: 'HeLLo_World_123'

```python
input='HeLLO_World_123'
# Uppercase Letter 
uppercase_letters= [char for char in input if char.isupper()]
print(uppercase_letters)
result_str="".join(uppercase_letters)
print(result_str)
```

12. Construction company calculates land plots. Create a Rectangle class with length and width. Add methods for area() and perimeter(). Test with 5x10 plot.

```python
class Rectangle:

    def __init__(self,length, width):
        self.length=length
        self.width=width
    def area(self):
        print(f"The area of Rectangle: {self.length*self.width}")
    def perimeter(self):
        print(f"The perimeter of Rectangle: {2*(self.length+self.width)}")

obj=Rectangle(5,10)
obj.area()
obj.perimeter()
```

13. Bank creates account management system. Build BankAccount class with holder name and balance. Implement deposit() and withdraw(). Perform: deposit 1000, withdraw 200, deposit 500.

```python
class BankAccount:
    def __init__(self,name, balance=0):
        self.name=name
        self.balance=balance
    # Method to Deposit
    def deposit(self, amount):
        if amount>0:
            self.balance+=amount
            print(f"Deposited ${amount}. New balance: ${self.balance}")
        else:
            print("Amount must be positive")

    # Method to Withdraw
    def withdraw(self, amount):
        if amount>0:
            if amount<=self.balance:
                self.balance-=amount
                print(f"Withdraw ${amount}. New balance: ${self.balance}")
            else: 
                print("Insufficient funds for this withdrawal.")
        else:
            print("Withdrawal amount must be positive.")

account=BankAccount("Naveen")

account.deposit(1000)
account.withdraw(200)
account.deposit(500)
```

19. Electronics store finds premium item. Given product: price dict, find the most expensive item. Input: {'laptop': 1200, 'mouse': 45, 'keyboard': 120, 'monitor': 350, 'headphones': 95}

```python
products={'laptop':1200,'mouse':45,'keyboard':120,'monitor':350,'headphones':95}
# Find the item with the highest price
most_expensive_item = max(products, key=products.get)
highest_price = products[most_expensive_item]

print(f"The most expensive item is the {most_expensive_item} at ${highest_price}.")
```

20. HR consolidates employee databases. Merge two employee ID: name dicts. dict1: {101: 'Alice', 102: 'Bob'}, dict2: {103: 'Charlie', 104: 'Diana'}. Output merged dict.

```python
dict1={101:'Alice',102:'Bob'}
dict2={103:'Charlie',104:'Diana'}

# Method 1: Using the merge operator `|` (Python 3.9+)
# Creates a new dictionary
merged_dict_1 = dict1 | dict2

# Method 2: Using dictionary unpacking `**` (Python 3.5+)
# Creates a new dictionary
merged_dict_2 = {**dict1, **dict2}

# Method 3: Using the `.update()` method
# Modifies dict1 in-place (does not create a new dictionary)
dict1.update(dict2)

print("Method 1:", merged_dict_1)
print("Method 2:", merged_dict_2)
print("Method 3:", dict1)
```

## Best Practices

1. **Test One Thing**: Each test should verify one specific behavior
2. **Clear Names**: Use descriptive names for test functions (e.g., `test_add_positive_numbers`)
3. **Arrange-Act-Assert**: Organize tests into setup, execution, and assertion phases
4. **DRY**: Use fixtures to avoid code duplication
5. **Test Edge Cases**: Test with empty inputs, large values, and error conditions
6. **Isolate Tests**: Tests should be independent and not rely on execution order
7. **Use Assertions**: PyTest assertions are cleaner and provide better failure messages

### Example: Well-structured test
```python
def test_divide_by_positive_number():
    # Arrange
    dividend, divisor, expected = 10, 2, 5
    
    # Act
    result = divide(dividend, divisor)
    
    # Assert
    assert result == expected, f"Expected {expected}, got {result}"
```

## 5. Fixtures in PyTest

Fixtures provide reusable setup and teardown code for tests:

````python
# Assertion examples
def test_assertions():
    """Demonstrate various assertion types"""
    # Basic assertions
    assert 1 == 1
    assert "hello" == "hello"
    assert [1, 2, 3] == [1, 2, 3]
    
    # Assertions with messages
    assert 5 > 3, "5 should be greater than 3"
    assert 10 in range(20), "10 should be in range"
    
    # Testing truthiness
    assert True
    assert [1, 2, 3]  # Non-empty list is truthy
    assert ""  # This would fail - empty string is falsy

# Test for exceptions
def test_exception_handling():
    """Test that functions raise expected exceptions"""
    with pytest.raises(ValueError):
        divide(10, 0)
    
    # You can also check exception message
    with pytest.raises(ValueError, match="Cannot divide"):
        divide(5, 0)
````

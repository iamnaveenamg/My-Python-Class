## 1. Classes and Objects

### Concept Overview
A **class** is a blueprint or template for creating objects. It defines the structure (attributes) and behavior (methods) of objects. An **object** is an instance of a class - a concrete realization of the blueprint.

**Key Concepts:**
- **Attributes**: Variables that store data for an object
- **Methods**: Functions that define the behavior of an object
- **Constructor (`__init__`)**: A special method that runs when an object is created
- **`self` parameter**: Refers to the current instance of the class

### Benefits:
- Organize code into logical, reusable units
- Bundle data and functions together
- Make code more modular and maintainable

```python
# Example 1: Basic Class Definition
class Car:
    """A class to represent a car."""
    
    def __init__(self, make, model, year):
        """Constructor: initialize car attributes"""
        self.make = make      # attribute
        self.model = model    # attribute
        self.year = year      # attribute
    
    def display_info(self):
        """Method: display car information"""
        print(f"{self.year} {self.make} {self.model}")
    
    def start_engine(self):
        """Method: simulate starting the engine"""
        return f"The {self.make} engine is now running!"

# Create objects (instances) from the Car class
car1 = Car("Toyota", "Camry", 2023)
car2 = Car("Honda", "Civic", 2022)

# Access attributes
print(f"Car 1: {car1.make} {car1.model}")
print(f"Car 2: {car2.year}")

# Call methods
car1.display_info()
print(car1.start_engine())
```

```python
# Example 2: Class with More Methods and Attributes
class Student:
    """A class to represent a student with grades."""
    
    # Class variable (shared by all instances)
    school_name = "Python Academy"
    
    def __init__(self, name, age, student_id):
        """Initialize student attributes"""
        self.name = name
        self.age = age
        self.student_id = student_id
        self.grades = []  # List to store grades
    
    def add_grade(self, grade):
        """Add a grade to the student's record"""
        self.grades.append(grade)
        print(f"Grade {grade} added for {self.name}")
    
    def calculate_average(self):
        """Calculate average grade"""
        if len(self.grades) == 0:
            return 0
        return sum(self.grades) / len(self.grades)
    
    def get_info(self):
        """Return student information"""
        avg = self.calculate_average()
        return f"{self.name} (ID: {self.student_id}), Age: {self.age}, Average: {avg:.2f}"

# Create student objects
student1 = Student("Alice", 20, "S001")
student2 = Student("Bob", 21, "S002")

# Add grades
student1.add_grade(85)
student1.add_grade(90)
student1.add_grade(88)

student2.add_grade(92)
student2.add_grade(95)

# Display information
print(student1.get_info())
print(student2.get_info())
print(f"School: {Student.school_name}")  # Accessing class variable
```

## 2. Inheritance

### Concept Overview
**Inheritance** allows a class (child/derived class) to inherit attributes and methods from another class (parent/base class). This enables code reuse and creates a hierarchical relationship between classes.

**Key Concepts:**
- **Parent Class (Base Class)**: The class being inherited from
- **Child Class (Derived Class)**: The class that inherits from the parent
- **Method Overriding**: Redefining a parent method in the child class
- **super()**: A function to call methods from the parent class

**Benefits:**
- Avoid code duplication
- Create logical hierarchy
- Extend functionality of existing classes
- Promote code reusability

```python
# Example 1: Basic Inheritance
class Animal:
    """Parent class: Animal"""
    
    def __init__(self, name, species):
        self.name = name
        self.species = species
    
    def speak(self):
        """Generic speak method"""
        return f"{self.name} makes a sound"
    
    def display_info(self):
        return f"Name: {self.name}, Species: {self.species}"

# Child classes that inherit from Animal
class Dog(Animal):
    """Child class: Dog"""
    
    def speak(self):
        """Override parent's speak method"""
        return f"{self.name} barks: Woof! Woof!"

class Cat(Animal):
    """Child class: Cat"""
    
    def speak(self):
        """Override parent's speak method"""
        return f"{self.name} meows: Meow! Meow!"

class Bird(Animal):
    """Child class: Bird"""
    
    def speak(self):
        """Override parent's speak method"""
        return f"{self.name} chirps: Tweet! Tweet!"

# Create objects and demonstrate inheritance
dog = Dog("Buddy", "Canine")
cat = Cat("Whiskers", "Feline")
bird = Bird("Tweety", "Avian")

print(dog.display_info())  # Uses parent's method
print(dog.speak())          # Uses child's overridden method

print("\n" + cat.display_info())
print(cat.speak())

print("\n" + bird.display_info())
print(bird.speak())
```

```python
# Example 2: Using super() to Call Parent Methods
class Vehicle:
    """Parent class: Vehicle"""
    
    def __init__(self, brand, color):
        self.brand = brand
        self.color = color
    
    def start(self):
        return f"Starting the {self.color} {self.brand} vehicle..."

class Motorcycle(Vehicle):
    """Child class: Motorcycle"""
    
    def __init__(self, brand, color, engine_cc):
        super().__init__(brand, color)  # Call parent's __init__
        self.engine_cc = engine_cc
    
    def start(self):
        # Call parent's start method and extend it
        parent_start = super().start()
        return f"{parent_start} Engine: {self.engine_cc}cc"

class Bicycle(Vehicle):
    """Child class: Bicycle"""
    
    def __init__(self, brand, color, num_gears):
        super().__init__(brand, color)
        self.num_gears = num_gears
    
    def start(self):
        parent_start = super().start()
        return f"{parent_start} Gears: {self.num_gears}"

# Create and test objects
motorcycle = Motorcycle("Harley-Davidson", "Black", 1200)
bicycle = Bicycle("Trek", "Red", 21)

print(motorcycle.start())
print(f"Details: {motorcycle.brand}, {motorcycle.color}, {motorcycle.engine_cc}cc\n")

print(bicycle.start())
print(f"Details: {bicycle.brand}, {bicycle.color}, {bicycle.num_gears} gears")
```

## 3. Encapsulation

### Concept Overview
**Encapsulation** is the bundling of data (attributes) and methods that operate on that data into a single unit (class), while hiding the internal details. It restricts direct access to some of an object's components, which is a means of preventing unauthorized access and modification.

**Key Concepts:**
- **Data Hiding**: Making attributes private to prevent direct external access
- **Private Attributes**: Prefixed with underscore `_` (convention) or double underscore `__` (name mangling)
- **Getters and Setters**: Methods to control access to private attributes
- **@property Decorator**: Controls how attributes are accessed and modified

**Benefits:**
- Control how data is accessed and modified
- Protect object integrity
- Hide internal complexity
- Implement validation logic

```python
# Example 1: Private Attributes and Getter/Setter Methods
class BankAccount:
    """A bank account class with encapsulation."""
    
    def __init__(self, account_holder, balance):
        self.__account_holder = account_holder  # Private attribute (double underscore)
        self.__balance = balance                 # Private attribute
    
    # Getter method
    def get_balance(self):
        """Return the account balance"""
        return self.__balance
    
    # Getter method
    def get_account_holder(self):
        """Return the account holder name"""
        return self.__account_holder
    
    # Setter method with validation
    def deposit(self, amount):
        """Deposit money with validation"""
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ${amount}. New balance: ${self.__balance}")
        else:
            print("Deposit amount must be positive!")
    
    def withdraw(self, amount):
        """Withdraw money with validation"""
        if amount > 0 and amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrawn ${amount}. New balance: ${self.__balance}")
        else:
            print("Invalid withdrawal amount!")

# Create account and use it
account = BankAccount("John Doe", 1000)

print(f"Account Holder: {account.get_account_holder()}")
print(f"Initial Balance: ${account.get_balance()}")

account.deposit(500)
account.withdraw(200)
account.withdraw(2000)  # Try to withdraw more than available

# Attempting direct access to private attributes
try:
    print(account.__balance)  # This will cause an error
except AttributeError as e:
    print(f"Error: Cannot access private attribute directly")
```

```python
# Example 2: Using @property and @setter Decorators
class Temperature:
    """A class to manage temperature with property decorators."""
    
    def __init__(self, celsius):
        self._celsius = celsius
    
    @property
    def celsius(self):
        """Get temperature in Celsius"""
        return self._celsius
    
    @celsius.setter
    def celsius(self, value):
        """Set temperature in Celsius with validation"""
        if value < -273.15:
            raise ValueError("Temperature cannot be below absolute zero (-273.15°C)")
        self._celsius = value
    
    @property
    def fahrenheit(self):
        """Get temperature in Fahrenheit (computed property)"""
        return (self._celsius * 9/5) + 32
    
    @fahrenheit.setter
    def fahrenheit(self, value):
        """Set temperature using Fahrenheit"""
        self._celsius = (value - 32) * 5/9
    
    @property
    def kelvin(self):
        """Get temperature in Kelvin (computed property)"""
        return self._celsius + 273.15

# Create temperature object and use property decorators
temp = Temperature(25)

print(f"Temperature: {temp.celsius}°C")
print(f"In Fahrenheit: {temp.fahrenheit}°F")
print(f"In Kelvin: {temp.kelvin}K")

# Modify using property setter
print("\nChanging temperature to 100°C...")
temp.celsius = 100
print(f"Temperature: {temp.celsius}°C")
print(f"In Fahrenheit: {temp.fahrenheit}°F")

# Modify using Fahrenheit setter
print("\nChanging temperature to 32°F...")
temp.fahrenheit = 32
print(f"Temperature: {temp.celsius}°C")
print(f"In Fahrenheit: {temp.fahrenheit}°F")

# Validation example
print("\nTrying to set invalid temperature...")
try:
    temp.celsius = -300
except ValueError as e:
    print(f"Error: {e}")
```

## 4. Polymorphism

### Concept Overview
**Polymorphism** means "many forms" - it's the ability of an object to take on many forms or for methods to behave differently based on the object they're called on. It allows you to write code that can work with objects of different types.

**Key Concepts:**
- **Method Overriding**: Child classes override parent methods with different implementations
- **Duck Typing**: "If it walks like a duck and quacks like a duck, it's a duck" - objects with the same interface can be used interchangeably
- **Operator Overloading**: Redefining operators like `+`, `-`, `*`, etc., for custom classes
- **Function Overloading**: Using default arguments and variable-length arguments

**Benefits:**
- Write flexible and reusable code
- Work with different types using the same interface
- Make code extensible without modification
- Reduce code duplication

```python
# Example 1: Polymorphism through Method Overriding
class Shape:
    """Parent class for shapes"""
    
    def area(self):
        raise NotImplementedError("Subclass must implement area method")
    
    def describe(self):
        return "This is a shape"

class Circle(Shape):
    """Circle class"""
    
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return 3.14159 * self.radius ** 2
    
    def describe(self):
        return f"Circle with radius {self.radius}"

class Rectangle(Shape):
    """Rectangle class"""
    
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height
    
    def describe(self):
        return f"Rectangle {self.width}x{self.height}"

class Triangle(Shape):
    """Triangle class"""
    
    def __init__(self, base, height):
        self.base = base
        self.height = height
    
    def area(self):
        return 0.5 * self.base * self.height
    
    def describe(self):
        return f"Triangle with base {self.base} and height {self.height}"

# Polymorphism in action - same method call, different behavior
shapes = [
    Circle(5),
    Rectangle(4, 6),
    Triangle(5, 9)
]

print("Polymorphism Example: Different shapes, same interface\n")
for shape in shapes:
    print(f"{shape.describe()}")
    print(f"Area: {shape.area():.2f}\n")
```

```python
# Example 2: Operator Overloading
class Vector:
    """A class to represent a 2D vector with operator overloading."""
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __add__(self, other):
        """Override + operator for vector addition"""
        return Vector(self.x + other.x, self.y + other.y)
    
    def __sub__(self, other):
        """Override - operator for vector subtraction"""
        return Vector(self.x - other.x, self.y - other.y)
    
    def __mul__(self, scalar):
        """Override * operator for scalar multiplication"""
        return Vector(self.x * scalar, self.y * scalar)
    
    def __eq__(self, other):
        """Override == operator for equality comparison"""
        return self.x == other.x and self.y == other.y
    
    def __str__(self):
        """String representation"""
        return f"Vector({self.x}, {self.y})"
    
    def magnitude(self):
        """Calculate vector magnitude"""
        return (self.x**2 + self.y**2)**0.5

# Create vectors and use overloaded operators
v1 = Vector(3, 4)
v2 = Vector(1, 2)

print(f"v1 = {v1}")
print(f"v2 = {v2}")

# Using overloaded operators
v3 = v1 + v2
print(f"\nv1 + v2 = {v3}")

v4 = v1 - v2
print(f"v1 - v2 = {v4}")

v5 = v1 * 2
print(f"v1 * 2 = {v5}")

print(f"\nv1 == v2: {v1 == v2}")
print(f"v1 == Vector(3, 4): {v1 == Vector(3, 4)}")

print(f"\nMagnitude of v1: {v1.magnitude():.2f}")
```

## 5. Decorators

### Concept Overview
A **decorator** is a function that takes another function or class as input and extends its behavior without permanently modifying the original. Decorators wrap a function or class with additional functionality.

**Key Concepts:**
- **Function Decorators**: Wrap functions to modify their behavior
- **Class Decorators**: Modify class definition
- **@staticmethod**: Method that doesn't require an instance
- **@classmethod**: Method that receives the class as first argument
- **@property**: Convert method to attribute (already covered)
- **Decorator with Arguments**: Decorators that accept parameters

**Benefits:**
- Add functionality to existing functions without modifying them
- Clean code by separating concerns
- Reuse decorator logic across multiple functions
- Reduce code duplication

```python
# Example 1: Simple Function Decorator
def my_decorator(func):
    """A simple decorator that prints function execution info"""
    def wrapper(*args, **kwargs):
        print(f"Function '{func.__name__}' is being called...")
        result = func(*args, **kwargs)
        print(f"Function '{func.__name__}' execution completed.")
        return result
    return wrapper

@my_decorator
def greet(name):
    """A simple greeting function"""
    print(f"Hello, {name}!")
    return f"Greeting {name}"

# Call the decorated function
result = greet("Alice")
print(f"Result: {result}\n")

# Another example with a calculator function
@my_decorator
def add(a, b):
    """Add two numbers"""
    return a + b

sum_result = add(5, 3)
print(f"Result: {sum_result}")
```

```python
# Example 2: Decorator with Arguments
def repeat(times):
    """Decorator that repeats function execution"""
    def decorator(func):
        def wrapper(*args, **kwargs):
            for i in range(times):
                print(f"Execution {i+1}:")
                result = func(*args, **kwargs)
            return result
        return wrapper
    return decorator

@repeat(times=3)
def say_hello(name):
    print(f"  Hello, {name}!")

print("Repeating function 3 times:\n")
say_hello("Bob")
```

```python
# Example 3: @staticmethod and @classmethod Decorators
class MathUtils:
    """Class demonstrating staticmethod and classmethod"""
    
    pi = 3.14159
    version = "1.0"
    
    @staticmethod
    def add(a, b):
        """Static method - doesn't need instance or class"""
        return a + b
    
    @staticmethod
    def multiply(a, b):
        """Another static method"""
        return a * b
    
    @classmethod
    def get_version(cls):
        """Class method - receives class as first argument"""
        return f"MathUtils version {cls.version}"
    
    @classmethod
    def create_from_radius(cls, radius):
        """Class method that acts as alternative constructor"""
        area = cls.pi * radius ** 2
        return f"Circle area with radius {radius}: {area:.2f}"

# Using static methods (don't require instance)
print("Static Methods:")
print(f"5 + 3 = {MathUtils.add(5, 3)}")
print(f"5 * 3 = {MathUtils.multiply(5, 3)}")

# Using class methods
print("\nClass Methods:")
print(MathUtils.get_version())
print(MathUtils.create_from_radius(4))
```

```python
# Example 4: Practical Decorator - Function Timer
import time

def timer_decorator(func):
    """Decorator that measures function execution time"""
    def wrapper(*args, **kwargs):
        start_time = time.time()
        print(f"Starting execution of '{func.__name__}'...")
        result = func(*args, **kwargs)
        end_time = time.time()
        execution_time = end_time - start_time
        print(f"Execution completed in {execution_time:.6f} seconds")
        return result
    return wrapper

@timer_decorator
def slow_function():
    """A function that takes time to execute"""
    time.sleep(1)
    return "Done!"

@timer_decorator
def fast_function():
    """A function that executes quickly"""
    total = sum(range(1000000))
    return f"Sum calculated: {total}"

# Test decorated functions
print("Test 1: Slow Function")
result1 = slow_function()
print(f"Result: {result1}\n")

print("Test 2: Fast Function")
result2 = fast_function()
print(f"Result: {result2}")
```

## Comprehensive Example: E-Commerce Order System

This example combines all OOP concepts: classes, inheritance, encapsulation, polymorphism, and decorators in a real-world scenario.

```python
# Comprehensive Example: E-Commerce Order System
from datetime import datetime
from functools import wraps

# Decorator for logging
def log_action(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        print(f"[{timestamp}] {func.__name__} called")
        result = func(*args, **kwargs)
        print(f"[{timestamp}] {func.__name__} completed\n")
        return result
    return wrapper

# Base Product class
class Product:
    """Base class for all products"""
    
    def __init__(self, name, price):
        self._name = name      # Encapsulation: private attribute
        self._price = price
    
    @property
    def name(self):
        return self._name
    
    @property
    def price(self):
        return self._price
    
    def get_discount_price(self, discount_percent):
        """Virtual method - to be overridden by subclasses"""
        return self._price * (1 - discount_percent / 100)
    
    def __str__(self):
        return f"{self._name} (${self._price:.2f})"

# Child classes demonstrating inheritance and polymorphism
class ElectronicsProduct(Product):
    """Electronics product with warranty (Inheritance)"""
    
    def __init__(self, name, price, warranty_months):
        super().__init__(name, price)
        self.warranty_months = warranty_months
    
    def get_discount_price(self, discount_percent):  # Polymorphism
        """Override parent method"""
        discounted = self._price * (1 - discount_percent / 100)
        # Add warranty fee
        warranty_fee = self._price * 0.05 * (self.warranty_months / 12)
        return discounted + warranty_fee

class BooksProduct(Product):
    """Book product (Inheritance)"""
    
    def __init__(self, name, price, author):
        super().__init__(name, price)
        self.author = author
    
    def get_discount_price(self, discount_percent):  # Polymorphism
        """Books get better discounts"""
        return self._price * (1 - (discount_percent + 10) / 100)

# Order class
class Order:
    """Order class with encapsulation"""
    
    _order_count = 0  # Class variable
    
    def __init__(self, customer_name):
        self.__customer_name = customer_name  # Private attribute
        self.__items = []
        self.__total = 0
        Order._order_count += 1
        self.__order_id = Order._order_count
    
    @property
    def order_id(self):
        """Getter for order ID"""
        return self.__order_id
    
    @property
    def customer_name(self):
        """Getter for customer name"""
        return self.__customer_name
    
    @log_action
    def add_item(self, product, quantity):
        """Add item to order"""
        self.__items.append({
            'product': product,
            'quantity': quantity,
            'price': product.price * quantity
        })
        self.__total += product.price * quantity
        print(f"  Added {quantity} x {product.name}")
    
    @log_action
    def apply_discount(self, discount_percent):
        """Apply discount to order"""
        discount_amount = self.__total * (discount_percent / 100)
        self.__total -= discount_amount
        print(f"  Discount: ${discount_amount:.2f}")
        print(f"  New total: ${self.__total:.2f}")
    
    @property
    def total(self):
        """Getter for total"""
        return self.__total
    
    @staticmethod
    def get_total_orders():
        """Static method to get total orders"""
        return Order._order_count
    
    @log_action
    def display_order(self):
        """Display order details"""
        print(f"Order ID: {self.__order_id}")
        print(f"Customer: {self.__customer_name}")
        print("Items:")
        for item in self.__items:
            print(f"  - {item['product']}: {item['quantity']} x ${item['price']:.2f}")
        print(f"Total: ${self.__total:.2f}")

# Create products (Polymorphism example)
laptop = ElectronicsProduct("Laptop", 999.99, 24)
book = BooksProduct("Python Programming", 39.99, "John Doe")
monitor = ElectronicsProduct("Monitor", 299.99, 12)

# Create order and add items
order = Order("Alice Smith")
order.add_item(laptop, 1)
order.add_item(book, 2)
order.add_item(monitor, 1)

# Apply discount and display
order.apply_discount(10)
order.display_order()

# Static method usage
print(f"Total orders processed: {Order.get_total_orders()}")
```

## Key Takeaways

### 1. **Classes and Objects**
- Classes define blueprints; objects are instances
- Use `__init__()` constructor for initialization
- Use `self` to reference instance attributes and methods

### 2. **Inheritance**
- Enables code reuse through class hierarchies
- Child classes inherit from parent classes
- Use `super()` to call parent class methods
- Override methods to customize behavior

### 3. **Encapsulation**
- Hide internal details using private attributes (`_attribute`, `__attribute`)
- Use getters and setters to control access
- Use `@property` decorator for cleaner syntax
- Protects object integrity through validation

### 4. **Polymorphism**
- Same method name, different implementations
- Allows flexible, reusable code
- Duck typing: "If it looks like an X, treat it like an X"
- Operator overloading customizes built-in operations

### 5. **Decorators**
- Wrap functions/methods to add functionality
- `@staticmethod`: method without instance/class access
- `@classmethod`: method with class access
- `@property`: convert methods to attributes
- Create custom decorators for cross-cutting concerns

### Best Practices
- Use meaningful class and method names
- Keep classes focused on a single responsibility
- Use encapsulation to protect data
- Leverage inheritance for code reuse, not forced hierarchies
- Use decorators to keep code DRY (Don't Repeat Yourself)
- Document your classes and methods with docstrings

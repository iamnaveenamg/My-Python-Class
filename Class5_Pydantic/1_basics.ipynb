```python
from pydantic import BaseModel
```

```python
class employee_schema(BaseModel):
    name:str
    id:int
    department:str
```

```python
obj=employee_schema(name="Naveen", id=120, department='Software Dev')
obj
```

```python
obj2 = employee_schema(**{"name": "John Doe", "id": 123, "department": "HR"})
obj2
```

Coercion VS Strict Types

```python
## Coercion
obj3 = employee_schema(**{"name": "John Doe", "id": "123", "department": "HR"})
obj3
```

```python
## Strict Type

from pydantic import StrictInt

class employee_schema(BaseModel):
    name:str
    id:int
    department:str

obj4 = employee_schema(**{"name": "John Doe", "id": 123, "department": "HR"})
obj4
```
